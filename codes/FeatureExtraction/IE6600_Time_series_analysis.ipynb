```python
from ts2vg import NaturalVG
from ts2vg import HorizontalVG
import numpy as np
```

```python
def nvg_re(method,country,df):
  result_list1 = []
  i = 1
  for i in range(2,df.shape[1]):
    g = NaturalVG()
    g.build(df.iloc[:,i])
    ig_g = g.as_igraph()
    result_list1.append([method,df.columns[i],np.mean(ig_g.degree()),ig_g.diameter(),ig_g.average_path_length(),country])
    i = i+1
  df1 = pd.DataFrame(result_list1,columns = ['Method','indicator code','Average degree','Network diameter','Average path length','country'])
  return df1
```

```python
def hvg_re(method,country,df):
  result_list1 = []
  i = 1
  for i in range(2,df.shape[1]):
    g = HorizontalVG()
    g.build(df.iloc[:,i])
    ig_g = g.as_igraph()
    result_list1.append([method,df.columns[i],np.mean(ig_g.degree()),ig_g.diameter(),ig_g.average_path_length(),country])
    i = i+1
  df1 = pd.DataFrame(result_list1,columns = ['Method','indicator code','Average degree','Network diameter','Average path length','country'])
  return df1
```

```python
empty_list = []
for id,df in df_cleaned_data.groupby(by = 'level_0'):
  empty_list.append(nvg_re('NVG',id,df))
empty_list2 = []
for id,df in df_cleaned_data.groupby(by = 'level_0'):
  empty_list2.append(hvg_re('HVG',id,df))
```

```python
df_nvg = pd.concat(empty_list,axis = 0)
df_hvg = pd.concat(empty_list2,axis = 0)
```

```python
df_nvg = df_nvg.dropna()
df_hvg = df_hvg.dropna()
```
